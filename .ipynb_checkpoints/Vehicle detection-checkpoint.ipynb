# Vehicle detection pipeline

This notebook applies a trained classifier to vehicle detection on a video file

## Load trained classifier

```python
import pickle

with open('classifier.pkl', 'rb') as f:
    scaler = pickle.load(f)
    svc = pickle.load(f)
    parameters = pickle.load(f)
```

## Run sliding window search on test images

```python
# Load a test images
import matplotlib.image as mpimg
import fnmatch
import os

test_imgs = []
folder = 'test_images'
for file in os.listdir(folder):
    if fnmatch.fnmatch(file, '*.jpg'):
        test_imgs.append(mpimg.imread(os.path.join(folder, file)))
        
print('{0} test images loaded'.format(len(test_imgs)))
```

```python
# Import function to find cars
from features import find_cars

# Define function to draw boxes on image
def draw_boxes(img, bboxes):
    draw_img = np.copy(img)
    for box in bboxes:
        cv2.rectangle(draw_img, box[0], box[1], (0, 0, 255), 6)
    return draw_img
```

```python
# Search parameters
ystart = 400
ystop = 656
scales = [1.0, 1.5, 2.0, 2.5, 3.0]
```

```python
# Plot a grid of test images with different scales
fig = plt.figure(figsize=(15, 10))

for im_idx, img in enumerate(test_imgs):
    for sc_idx, scale in enumerate(scales):
        # Add subplot
        ax = fig.add_subplot(len(test_imgs), len(scales), im_idx*len(scales) + sc_idx + 1)
        # Find cars
        boxes = find_cars(img, ystart, ystop, scale, svc, scaler,
                          hog_channel=parameters['hog_channel'], 
                          spatial_size=(parameters['spatial'], parameters['spatial']),
                          hist_bins=parameters['histbin'],
                          color_space=parameters['color_space'])
        # Annotate image with boxes
        ax.imshow(draw_boxes(img, boxes))
        ax.set_axis_off()
        # Label scales
        if im_idx == 0:
            ax.set_title('Scale: {0}'.format(scale))
fig.tight_layout()

fig.savefig('output_images/sliding_windows.png')
```

## Implement heatmap and false positive rejection

```python
def process_img(img, scales, threshold):
    # Get boxes for each of the scales
    bboxes = []
    for scale in scales:
        scale_boxes = find_cars(img, ystart, ystop, scale, svc, scaler,
                                hog_channel=parameters['hog_channel'], 
                                spatial_size=(parameters['spatial'], parameters['spatial']),
                                hist_bins=parameters['histbin'],
                                color_space=parameters['color_space'])
        bboxes = bboxes + scale_boxes
    # Create heatmap
    heatmap = np.zeros(img[:,:,0].shape)
    for box in bboxes:
        # Add += 1 for all pixels inside each bbox
        heatmap[box[0][1]:box[1][1], box[0][0]:box[1][0]] += 1

    # Return thresholded heatmap
    heatmap[heatmap <= threshold] = 0
    return heatmap
```

```python
# Plot heatmaps for each test image
thresh = 1
scales = [1.0, 1.5, 2.0]

fig = plt.figure(figsize=(15, 3))

for im_idx, img in enumerate(test_imgs):
    # Add subplot for image
    ax = fig.add_subplot(2, len(test_imgs), im_idx + 1)
    ax.imshow(img)
    ax.set_axis_off()
    # Add subplot for heatmap
    ax = fig.add_subplot(2, len(test_imgs), im_idx + 1 + len(test_imgs))
    # Process image
    heatmap = process_img(img, scales, thresh)
    # Annotate image with boxes
    ax.imshow(heatmap)
    ax.set_axis_off()
fig.tight_layout()

fig.savefig('output_images/heatmaps.png')
```

## Class to implement pipeline on video

```python
from collections import deque
from scipy.ndimage.measurements import label
```

```python
class CarDetector():
    def __init__(self, **kwargs):
        # Store parameters for future use
        self.settings = kwargs
        # Create circular buffer for storing boxes
        self.buffer = deque(maxlen=self.settings['frames'])
        # Variable to store previous frame
        self.previous_frame = None
        # Variable to store detected boxes
        self.boxes = None
    
    def labels_to_boxes(self, img, labels):
        # Iterate through all detected cars
        boxes = []
        for car_number in range(1, labels[1]+1):
            # Find pixels with each car_number label value
            nonzero = (labels[0] == car_number).nonzero()
            # Identify x and y values of those pixels
            nonzeroy = np.array(nonzero[0])
            nonzerox = np.array(nonzero[1])
            # Define a bounding box based on min/max x and y
            bbox = ((np.min(nonzerox), np.min(nonzeroy)), (np.max(nonzerox), np.max(nonzeroy)))
            boxes.append(bbox)
        # Return the boxes
        return boxes
        
    def annotate_img(self, img, boxes):
        # Iterate through all detected cars
        for bbox in boxes:
            # Draw the box on the image
            cv2.rectangle(img, bbox[0], bbox[1], (0, 0, 255), 6)
        # Text annotation
        cv2.putText(img, 'Vehicles detected: {0}'.format(len(boxes)), 
                    (20, 50), cv2.FONT_HERSHEY_SIMPLEX, 1, (0, 255, 0), 2)
        # Return the image
        return img
        
    def update(self, img):
        # Find cars in the image
        bboxes = []
        for scale in self.settings['scales']:
            scale_boxes = find_cars(img, 
                                    self.settings['ystart'], self.settings['ystop'],
                                    scale, 
                                    self.settings['svc'], self.settings['scaler'],
                                    hog_channel=self.settings['hog_channel'], 
                                    spatial_size=(self.settings['spatial'], self.settings['spatial']),
                                    hist_bins=self.settings['histbin'],
                                    color_space=self.settings['color_space'])
            bboxes = bboxes + scale_boxes
        # Create heatmap from list of boxes
        heatmap = np.zeros(img[:,:,0].shape)
        for box in bboxes:
            # Add += 1 for all pixels inside each bbox
            heatmap[box[0][1]:box[1][1], box[0][0]:box[1][0]] += 1
        # Add heatmap to buffer
        self.buffer.append(heatmap)
        
        # Don't start annotating images until buffer is full
        if len(self.buffer) < self.settings['frames']:
            return img
        # Threshold heatmaps in buffer
        for frame in self.buffer:
            heatmap = np.sum(self.buffer, axis=0)
        heatmap[heatmap <= self.settings['threshold']] = 0
        # Label the heatmap
        labels = label(heatmap)
        # Convert to boxes
        boxes = self.labels_to_boxes(img, labels)
        self.boxes = boxes        
        # Store previous frame
        self.previous_frame = img
        # Return annotated image
        return self.annotate_img(img, boxes)
```

## Process video files

```python
# Import everything needed to edit/save/watch video clips
from moviepy.editor import VideoFileClip
from IPython.display import HTML
```

```python
settings = {'ystart': 400, 'ystop': 656, 'scales': [1.0, 1.5, 2.0], **parameters,
            'frames': 20, 'threshold': 25,
            'svc': svc, 'scaler': scaler}
```

```python
in_file = 'project_video.mp4'
out_file = 'output_images/project_video.mp4'

clip = VideoFileClip(in_file)
car_detector = CarDetector(**settings)

clip_processed = clip.fl_image(car_detector.update)
%time clip_processed.write_videofile(out_file, audio=False)
```
